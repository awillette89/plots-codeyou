```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'HenName': ['Bella', 'Daisy', 'Luna', 'Rosie', 'Ruby', 'Molly', 'Lily', 'Coco', 'Penny', 'Hazel'],
    'EggsPerWeek': [6, 7, 5, 6, 6, 4, 7, 5, 6, 5],
    'AgeMonths': [18, 20, 16, 22, 19, 24, 21, 17, 23, 25],
    'FeedGramsPerDay': [120, 125, 118, 122, 124, 130, 123, 119, 126, 128]
}

hens = pd.DataFrame(data)
print(hens)
```

```python
#pandas bar plot

basic_plot = hens.plot(kind='bar', x='HenName', y='EggsPerWeek', title='Eggs Laid by Each Hen');
basic_plot.set_xlabel('Hen');
basic_plot.set_ylabel('Eggs per Week');
```

```python
#pandas bar plot sorted by most eggs laid

hens = hens.sort_values(by='EggsPerWeek', ascending=False)
basic_plot = hens.plot(kind='bar', x='HenName', y='EggsPerWeek', title='Eggs Laid by Each Hen');
basic_plot.set_xlabel('Hen');
basic_plot.set_ylabel('Eggs per Week');
```

```python
#matplot bar graph for feed to egg conversion by hen

hens['FeedConversion'] = hens['EggsPerWeek']/(hens['FeedGramsPerDay']*7)

plt.figure(figsize=(10, 6))
plt.bar(hens['HenName'], hens['FeedConversion'])

#labels and title
plt.title('Feed Conversion by Hen')
plt.ylabel('Egg per Gram of Feed')
plt.xlabel('Hen')
```

```python
#previous matplot graph sorted

hens_sorted = hens.sort_values(by=['FeedConversion'], ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(hens_sorted['HenName'], hens_sorted['FeedConversion'])

#labels and title
plt.title('Feed Conversion by Hen')
plt.ylabel('Egg per Gram of Feed')
plt.xlabel('Hen')
```

```python
#seaborn graph of sorted feed conversion

sns.set(style="darkgrid")

plt.figure(figsize=(10,6))
bar_plot = sns.barplot(x='HenName', y='FeedConversion', data=hens_sorted, dodge=False)

bar_plot.set_title('Feed Conversion by Hen', fontsize=16)
bar_plot.set_ylabel('Eggs per Gram of Feed', fontsize=12)
bar_plot.set_xlabel('Hen', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
#seaborn graph of sorted eggs per week

sns.set(style="darkgrid")

hens = hens.sort_values(by='EggsPerWeek', ascending=False)
plt.figure(figsize=(10,6))
bar_plot = sns.barplot(x='HenName', y='EggsPerWeek', data=hens, dodge=False)

bar_plot.set_title('Eggs Laid by Each Hen', fontsize=16)
bar_plot.set_ylabel('Eggs per Week', fontsize=12)
bar_plot.set_xlabel('Hen', fontsize=12)

plt.tight_layout()
plt.show()
```
