```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'HenName': ['Bella', 'Daisy', 'Luna', 'Rosie', 'Ruby', 'Molly', 'Lily', 'Coco', 'Penny', 'Hazel'],
    'EggsPerWeek': [6, 7, 5, 6, 6, 4, 7, 5, 6, 5],
    'AgeMonths': [18, 20, 16, 22, 19, 24, 21, 17, 23, 25],
    'FeedGramsPerDay': [120, 125, 118, 122, 124, 130, 123, 119, 126, 128]
}

hens = pd.DataFrame(data)
print(hens)
```

```python
#pandas bar plot

basic_plot = hens.plot(kind='bar', x='HenName', y='EggsPerWeek', title='Eggs Laid by Each Hen');
basic_plot.set_xlabel('Hen');
basic_plot.set_ylabel('Eggs per Week');
```

```python
#pandas bar plot sorted by most eggs laid

hens = hens.sort_values(by='EggsPerWeek', ascending=False)
basic_plot = hens.plot(kind='bar', x='HenName', y='EggsPerWeek', title='Eggs Laid by Each Hen');
basic_plot.set_xlabel('Hen');
basic_plot.set_ylabel('Eggs per Week');
```

```python
#matplot bar graph for feed to egg conversion by hen

feed_conversion = hens['EggsPerWeek']/(hens['FeedGramsPerDay']*7)

plt.figure(figsize=(10, 6))
plt.bar(hens['HenName'], feed_conversion)

#labels and title
plt.title('Feed Conversion by Hen')
plt.ylabel('Egg per Gram of Feed')
plt.xlabel('Hen')
```
